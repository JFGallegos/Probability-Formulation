# Probability Formulation

```python
import numpy as np
import matplotlib.pyplot as plot
```

## One dimension distribution

```python
x = np.random.randn(100000) ## Generate 100,000 samples of a Gaussian Distribution N(x|0,1)
dummy = plot.hist(x,100)
print('moyenne=', np.mean(x))
print('variance=', np.var(x))
```

```python
y = 2*x - 1 # Transform the sample to a linear form

dummy = plot.hist(y,100)
print ("mean=",np.mean(y))    # Should be arround 2*E[x] -1 = -1
print ("variance=",np.var(y))    # Shoud be arround 2**2 * var[x] = 4
```

## Two dimensional distribution

```python
## E.g., 1
mu = np.array([0,0])
sigma = np.eye(2)
x = np.random.multivariate_normal(mu,sigma,100000).T # Generate 100,000 samples of a Gaussian Distribution N(x|0,I) in two dimensions
dummy = plot.hist2d(x[0,:],x[1,:],50,cmap = plot.cm.Greys)
print ("mean=", np.mean(x,axis=1))
print ("covariance=\n", np.cov(x))

```

```python
## E.g., 2

mu = np.array([1,2])
sigma = np.array([[2,0],[0,1]])
y = np.random.multivariate_normal(mu,sigma,100000).T # Generate 100,000 samples of a Gaussian Distribution N(x|[1,2]^T,[[2,0],[0,1]]) in 2 dimensions
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = plot.cm.Greys)

print ("mean=",np.mean(y,axis=1))
print ("covariance=\n",np.cov(y))
```

```python
## E.g., 3

mu = np.array([1,2])
sigma = np.array([[1.5,1],[1,1.5]])
y = np.random.multivariate_normal(mu,sigma,100000).T # Generate 100,000 samples of a Gaussian Distribution N(x|[1,2]^T,[[1.5,1],[1,1.5]]) in 2 dimensions
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = plot.cm.Greys)

print ("moyenne=",np.mean(y,axis=1))
print ("covariance=\n",np.cov(y))
```

```python
## E.g., 4

mu = np.array([2,-1])
A = np.array([[3,0],[0,0.5]])
y = np.dot(A,x) + mu.reshape((2,1))
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = plot.cm.Greys)

print ("mean=",y.mean(axis=1))
print ('which is similar to the values of',mu)  # E[A x + mu] = A E[x]+mu = A(0)+mu = mu
print ("covariance=\n",np.cov(y))
print ("which is similar to the values of",np.dot(A,A.T))   # cov[A x + mu] = A cov[x] A^T = A I A^T = A A^T

```

```python
## E.g.5
mu = np.array([2,-1])
A = np.array([[2,1],[1,2]])
y = np.dot(A,x) + mu.reshape((2,1))
dummy = plot.hist2d(y[0,:],y[1,:],50,cmap = plot.cm.Greys)

print ("mean=",y.mean(axis=1))
print ('which is similar to the values of',mu)    # E[A x + mu] = A E[x]+mu = A(0)+mu = mu
print ("covariance=\n",np.cov(y))
print ('which is similar to the values of',np.dot(A,A.T))   # cov[A x + mu] = A cov[x] A^T = A I A^T = A A^T
```
